## Lecure number 4 Creating a Conditional Column from More Than 2 Choices: np.select()

```python
import numpy as np
```

```python

# create an array based on multiple price tiers (+2 choices)

# create "conditions" and "values" variables

conditions = [(df['Price_euros'] > 3000) , 
             (df['Price_euros'] > 2000) & (df['Price_euros'] <= 3000) , 
             (df['Price_euros'] > 800) & (df['Price_euros'] <= 2000) ,
             (df['Price_euros'] < 800)
            ]
choices = ['Too expensive' , 'Expensive' , 'Affordable' , 'Cheap']
```

```python
df["Price_tier"] = np.select(conditions, choices, default="Unknown")
```

```python
df
```

```python
df.value_counts('Price_tier')
```

```python

# create an array based on multiple screen size tiers (+2 choices)
# Too Big>16, Big>14, Small>12, Too Small<12
conditions = [(df['Inches'] > 16) , 
             (df['Inches'] > 12) & (df['Inches'] <= 16) , 
             (df['Inches'] > 12) & (df['Inches'] <= 15) ,
             (df['Inches'] < 12)
            ]
choices = ['Too Big', 'Big', 'Small', 'Too Small']
```

```python
df['Size comparision']  = np.select(conditions, choices , default ="Unknown")
```

```python
df.value_counts('Size comparision')
```
