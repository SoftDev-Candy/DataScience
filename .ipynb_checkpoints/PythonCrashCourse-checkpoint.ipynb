```python
def calculate_total(numbers):
    total = 0
    for i in numbers:
        total += i
    return total

print(calculate_total(expenses))
```

```python
expenses = [
    {"category": "rent", "amount": 1600},
    {"category": "food", "amount": 800},
    {"category": "transport", "amount": 150},
    {"category": "gym", "amount": 120},
    {"category": "food", "amount": 200}
]
total_spending = 0
food_total  = 0
highest_expense = 0
for item in expenses:
    total_spending += item["amount"]
    if(item["category"] == "food"):
        food_total  += item["amount"]
    if( highest_expense < item["amount"]):
            highest_expense = item["amount"]



print("Total Spending :" , total_spending)
print("Food spending: " , food_total)
print("Highest Expense: " , highest_expense)

    
    

```
