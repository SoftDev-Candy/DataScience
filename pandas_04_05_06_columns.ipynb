```python
df["total_score"] = df["math score"] + df["reading score"] + df["writing score"]
df.head()
```

```python
df["average_score"] = df["total_score"] /3 
df.head()
```

```python
df[["math score", "reading score", "writing score", "total_score", "average_score"]].head()
```

```python
df["passed_math"] = df["math score"] >= 50
```

```python
df["passed_math"].value_counts()
```

```python
means = pd.Series({
    "Math": df["math score"].mean(),
    "Reading": df["reading score"].mean(),
    "Writing": df["writing score"].mean()
})

print("Math Average:", means["Math"])
print("Reading Average:", means["Reading"])
print("Writing Average:", means["Writing"])

best_subject = means.idxmax()
best_avg = means.max()

print("Best Subject:", best_subject, f"({best_avg})")

```

```python
df["strong_student"] = df["average_score"] >= 80
df["strong_student"].value_counts()
```
