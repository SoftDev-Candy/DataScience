```python
# Calculate means and store in a Series
means = pd.Series({
    "Math": df_exams["math score"].mean(),
    "Reading": df_exams["reading score"].mean(),
    "Writing": df_exams["writing score"].mean()
})

# Print averages
print("Math Average:", means["Math"])
print("Reading Average:", means["Reading"])
print("Writing Average:", means["Writing"])

# Find the best subject name and its value
best_subject = means.idxmax()
best_avg = means.max()

print("Best Subject:", best_subject, f"({best_avg})")

```
